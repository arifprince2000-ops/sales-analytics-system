```python
import pandas as pd
```

```python
df = pd.read_csv("sales.csv", sep="|")
```

```python
df_clean = df.drop_duplicates(subset=['CustomerID','Region'])
```

```python
df_clean_1 = df_clean[df_clean['Quantity'] > 0]
```

```python
df_clean_1['UnitPrice'] = pd.to_numeric(df_clean_1['UnitPrice'], errors='coerce')
df_clean_2 = df_clean_1[df_clean_1['UnitPrice'] > 0].copy()
```

```python
df_clean_1['UnitPrice'] = pd.to_numeric(df_clean_1['UnitPrice'], errors='coerce')
```

```python
df_clean_2 = df_clean_1[df_clean_1['UnitPrice'] > 0]
```

```python
df_clean_3 = df_clean_2[df_clean_2['TransactionID'].str.startswith('T', na=False)]
```

```python
df_clean_3['ProductName'] = df_clean_3['ProductName'].str.replace(',', '', regex=False)
```

```python
df_clean_3.dropna(how='all')
```

```python
df_TotalSales = df_clean_3['Quantity'] * df_clean_3['UnitPrice']
```

```python
df_sumTotalSales = df_TotalSales.sum()
```

```python
print("Total_Sales",":", df_sumTotalSales)
```

```python
df_clean_3['TotalSales'] = df_clean_3['Quantity'] * df_clean_3['UnitPrice']
```

```python
region_sales = df_clean_3.groupby('Region')['TotalSales'].sum().sort_values(ascending=False)
```

```python
print(region_sales)
```

```python
transactions_per_region = df_clean_3.groupby('Region')['TransactionID'].nunique().sort_values(ascending=False)
print(transactions_per_region)
```

```python
region_pct = df_clean_3.groupby('Region')['TotalSales'].sum() / df_sumTotalSales * 100
region_pct = region_pct.sort_values(ascending=False).round(2)
print(region_pct)
```

```python
top_products = df_clean_3.groupby('ProductName').agg({
    'Quantity': 'sum',
    'TotalSales': 'sum'
}).round(2).nlargest(5, 'Quantity').reset_index()
top_products.columns = ['ProductName', 'TotalQuantity', 'TotalRevenue']
print(top_products)
```

```python
customer_summary = df_clean_3.groupby('CustomerID').agg(
    total_spent=('TotalSales', 'sum'),
    purchase_count=('TransactionID', 'nunique'),
    avg_order_value=('TotalSales', 'mean')
).round(2).sort_values('total_spent', ascending=False)

customer_summary['products_bought'] = [
    df_clean_3[df_clean_3['CustomerID'] == cid]['ProductName'].unique().tolist()
    for cid in customer_summary.index
]

customer_summary = customer_summary.reset_index()
all_customers_dict = customer_summary.to_dict('records')
```

```python
print(all_customers_dict)
```

```python
df_clean_3['Date'] = pd.to_datetime(df_clean_3['Date'])
```

```python
daily_trends = df_clean_3.groupby(df_clean_3['Date'].dt.date).agg(
    revenue=('TotalSales', 'sum'),
    transaction_count=('TransactionID', 'nunique'),
    unique_customers=('CustomerID', 'nunique')
).sort_index()
```

```python
daily_trends.index = daily_trends.index.astype(str)
daily_trends_dict = daily_trends.to_dict('index')
```

```python
print(daily_trends_dict)
```

```python
top_day_row = daily_trends.loc[daily_trends['revenue'].idxmax()]
```

```python
highest_revenue_details = (
    top_day_row.name, 
    top_day_row['revenue'].item(), 
    int(top_day_row['transaction_count'])
)
```

```python
print(highest_revenue_details)
```
